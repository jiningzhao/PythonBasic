```python
# 所谓封装是把对象的属性和行为结合在一个独立的系统单位内部
# 尽可能隐蔽对象的内部细节，只向外部提供接口
# 降低对象间的耦合度
# 封装的重要意义：
#     使对象能够集中
```

```python
# 继承，也称泛化，继承性是子类自动共享父类属性和方法的机制，在定义和实现一个类的时候，可以在一个已经存在的类的基础之上来进行，把这个已经存在的类所定义的内容
# 作为自己的内容，并加入自己若干新的内容
# 继承简化了人们对事物的认识和描述，有益于软件复用，是OO奇数提高
```

```python
class Person(object):                  # class是定义类的关键字
                                       # Person是类的名字，通常首字母大写
                                       # ()里面写父类的名字,object是所有类的基类，如果不写object表示基类是object
        def __init__(self,name):       # __init__是类的构造方法，在创建类的对象时调用此方法，第一个参数必须是self
            self.name=name             # 在构造方法内定义的属性self.name是实例属性
            print('init:',name)
        def getName(self):             # 类的普通方法，参数的第一个必须是self
            return self.name
        
# 创建一个实例
person1=Person('张三')                 # person1是类的一个实例，即对象
                                       # 类名（参数），参数需要按照__init__规定的格式写，第一个self不需要为其赋值
                                       # 构造出一个实例后，自动调用__init__方法 
name1=person1.getName()                # 实例化后，可以通过“对象名.方法名”的方式来访问类里的方法，参数同样不需要为self赋值
print(name1)

# 创建另一个实例
person2=Person('李四')
name2=person2.getName()
print(name2)
```

### 访问控制

```python
class A:
    def __init__(self,value1=0,value2=0):
        self._value1=value1   #_value1是保护属性，类本身、子类可用
        self.__value2=value2  #__value2是私有属性，类本身可用
    def setValue(self,value1,value2):
        self._value1=value1
        self.__value2=value2
    def __show(self):       # 在方法前面加下划线可以改变方法的访问属性
        print(self._value1)
        print(self.__value2)

a=A(1,2)                    # 构造类A的实例a
print(a._value1)            # 访问保护属性：实例名._属性名
print(a._A__value2)         #访问私有属性：实例名._类名__属性名
a._A__show()        
```

```python
# 类方法和静态方法
class Root:
    __total=0
    def __init__(self,v):
        self.__value=v         #实例属性，对象的成员
        Root.__total+=1        #类属性，类的成员
    def show(self):           #公有方法，对象的方法，可以访问实例属性和类属性
        print('self.__value:',self.__value)
        print('Root.__total',Root.__total)
    @classmethod              # 修饰类方法
    def classShowTotal(cls):       # cls为类方法的第一参数，类方法只能访问类的成员，不能访问对象的成员
        print("classmethod:",cls.__total)
    @staticmethod             #修饰静态方法
    def staticShowTotal():         #静态方法只能访问类的成员，不能访问对象的成员
        print("staticmethod:",Root.__total)
a1=Root(1)
a2=Root(3)
a1.show()
a2.show()
print(Root._Root__total)
#静态方法和类方法可通过类名、对象名调用
a1.classShowTotal()
Root.classShowTotal()
a1.staticShowTotal()
Root.staticShowTotal()
```

```python
class Score(object):
    __count=0
    __sumChinese=0
    __sumMath=0
    def __init__(self,name,chinese,math):
        self.__name=name
        self.__chinese=chinese
        self.__math=math
        Score.__count+=1
        Score.__sumChinese+=chinese
        Score.__sumMath+=math
    def CalcTotalScore(self):
        return self.__chinese+self.__math
    @staticmethod
    def CalcAverageScore(total,num):
        return total/num
    def printStudent(self):
        print("{name}:语文{c}分，数学{m}分，总分{s}分，平均分{a}分".format(name=self.__name,c=self.__chinese,m=self.__math,s=self.CalcTotalScore(),
                                                           a=self.CalcAverageScore(self.CalcTotalScore(),2)))
    @classmethod
    def printClass(cls):
        chAv=cls.CalcAverageScore(cls.__sumChinese,cls.__count)
        maAv=cls.CalcAverageScore(cls.__sumMath,cls.__count)
        print("班级总平均分{},语文平均分{},数学平均分{}".format(chAv+maAv,chAv,maAv))
s1=Score('张三',99,79)
s2=Score('李四',100,65)
s3=Score('王五',90,96)
s1.printStudent()
s2.printStudent()
s3.printStudent()
Score.printClass()
```

```python
#1.定义圆形类Circle,具有属性半径和颜色，具有方法设置颜色、设置半径、计算周长、计算面积。
#在主程序中创建一个圆形对象，半径为3，颜色为绿色，打印其颜色、周长和面积；再将其半径改为10，颜色改为红色，打印其颜色、周长和面积
class Circle(object):
    def __init__(self,r=0,color='红'):
        self.__r=r
        self.__color=color
    def setColor(self,color):
        self.__color=color
        return self.__color
    def setR(self,r):
        self.__r=r
        return self.__r
    def getC(self):
        return self.__r*2*3.14
    def getS(self):
        return (self.__r**2)*3.14
    def print_Circle(self):
        print('半径为{}的圆，颜色为{}色，周长为{}，面积为{}。'.format(self.__r,self.__color,self.getC(),self.getS()))
c=Circle(3,'绿')
c.print_Circle()
c.setR(10)
c.setColor('红')
c.print_Circle()
```

```python
#2.定义一个类Point，属性是点的x、y坐标，方法包括1）设置点的坐标；2）计算该点与原点的距离，
#3）计算以该点为圆心，以该点与原点的距离为半径的圆的周长和面积；4）计算该点和原点构成的长方形的周长和面积
#在主程序中创建对象，坐标为（3，4），打印以上结果，将坐标改为（-5，-8），打印以上计算结果
import math
class Point(object):
    def __init__(self,x,y):
        self.__x=x
        self.__y=y
    def setPoint(self,x,y):
        self.__x=x
        self.__y=y
        return (self.__x,self.__y)
    def getDistance(self):
        return math.sqrt(self.__x**2+self.__y**2)
    def getCircle_C_S(self):
        circle_C=self.getDistance()*2*3.14
        circle_S=(self.getDistance()**2)*3.14
        return circle_C,circle_S
    def getRectangle_C_S(self):
        rectangle_C=(abs(self.__x)+abs(self.__y)*2
        rectangle_S=abs(self.__x*self.__y)
        return rectangle_C,rectangle_S
    def printInfo(self):
        c1,s1=self.getCircle_C_S()
        c2,s2=self.getRectangle_C_S()
        print('点坐标{}，该点与原点的距离为{}，圆的周长为{}、面积为{}，长方形周长为{}、面积为{}'.format(self.setPoint(self.__x,self.__y),self.getDistance(),c1,s1,c2,s2))
point1=Point(3,4)
point1.printInfo()
point1.setPoint(-5,-8)
point1.printInfo()
        
```

```python
#练习3：定义一个时钟类Clock，属性是时，分，秒，用于完成时间的计算和显示，方法包括
#1）计算上面的时间之后指定m小时，n分钟，k秒之后的时间
#2）计算上面的时间之前指定m小时，n分钟，k秒之后的时间
#3）函数参数为时，分，秒，将传入的时分秒显示成“小时：分钟：秒AM”的格式，最后显示的AM、PM表示上下午
#在主程序中创建对象，初始时间是15:26:35，计算此时间之后的53:42：40的时间，此时间之前53:42:40的时间，并打印结果
class Clock(object):
    def __init__(self,hour,minute,second):
        self.__hour=hour
        self.__minute=minute
        self.__second=second
    def behindTime(self,m,n,k):
        s1=self.__second+k
        if s1>=60:
            self.__second=s1%60
            self.__minute+=s1//60
        s2=self.__minute+n
        if s2>=60:
            self.__minute=s2%60
            self.__hour+=s2//60
        s3=self.__hour+m
        if s3>=24:
            self.__hour=s3%24
        self.__M='PM'
        if self.__hour<12:
            self.__M='AM'
        else:
            self.__hour-=12
        return self.__hour,self.__minute,self.__second,self.__M
    def frontTime(self,m1,n1,k1):
        z1=self.__hour+24-m1%24
        if z1>=24:
            self.__hour=z1%24
        else:
            self.__hour=z1
        z2=self.__minute+60-n1
        if  z2>=60:
            self.__minute=z2%60
        else:
            self.__minute=z2
            self.__hour-=1
            if self.__hour<0:
                self.__hour=23
        z3=self.__second+60-k1
        if  z3>=60:
            self.__second=z3%60
        else:
            self.__second=z3
            self.__minute-=1
            if self.__minute<0:
                self.__minute=59
        self.__M='PM'
        if self.__hour<12:
            self.__M='AM'
        else:
            self.__hour-=12
    def printInfo(self):
        print('{:0>2d}:{:0>2d}:{:0>2d}.{}'.format(self.__hour,self.__minute,self.__second,self.__M))
time=Clock(15,26,35)
time.behindTime(53,42,40)
time.printInfo()
time1=Clock(15,26,35)
time1.frontTime(53,42,40)
time1.printInfo()
```

```python
class Clock(object):
    def __init__(self,h,m,s):
        self.__hour=h
        self.__min=m
        self.__sec=s
    @staticmethod
    def display(h,m,s):
        str1='AM'
        if h>12:
            str1='PM'
            h-=12
        else:
            str1='AM'
        print('{:0>2d}:{:0>2d}:{:0>2d}'.format(h,m,s)+str1)
    def After(self,h,m,s):
        second=(self.__sec+s)%60
        second_zheng=(self.__sec+s)//60    #需要向分钟进位部分
        minute=(self.__min+m+second_zheng)%60
        minute_zheng=(self.__min+m+second_zheng)//60
        hour=(self.__hour+h+minute_zheng)%24
        Clock.display(hour,minute,second)   #显示时间
    def Before(self,h,m,s):
        second=(self.__sec-s)%60
        second_zheng=(self.__sec-3)//60
        minute=(self.__min-m+second_zheng)%60
        minute_zheng=(self.__min-m+second_zheng)//60
        hour=(self.__hour-h+minute_zheng)%24
        Clock.display(hour,minute,second)
Clock.display(15,5,20)
time1=Clock(15,26,35)
time1.After(53,42,40)
time1.Before(53,42,40)
```

```python
#两个游戏里的拳击手对打，属性包括血的厚度，攻击力，防御力
#受到攻击后，血的厚度-=对方的攻击力-自己的防御力
#键盘输入攻击方的编号，如输入1则拳击手1攻击，输入2攻击手2攻击，每次攻击后，显示两个人剩余的血量
#当一方血量为0或负值时，游戏结束，打印胜利方
class Person(object):
    def __init__(self,blood,fight,defense):
        self.__blood=blood
        self.__fight=fight
        self.__defense=defense
    def beFight(self,fight):
        self.__blood-=fight-self.__defense
        return self.__blood
    def printInfo(self):
        print('剩余血量：',self.__blood)
p1=Person(100,25,20)
p2=Person(100,30,15)
print('p1的初始血量为：{}攻击力为：{}防御力为：{}'.format(p1._Person__blood,p1._Person__fight,p1._Person__defense))
print('p2的初始血量为：{}攻击力为：{}防御力为：{}'.format(p2._Person__blood,p2._Person__fight,p2._Person__defense))
while True:
    f=int(input('请输入攻击方：'))
    print('{}攻击！'.format(f))
    if f==1:
        p2.beFight(p1._Person__fight)
        print('p2',end='')
        p2.printInfo()
        if p2._Person__blood<=0:
            print('p2没血了，凉了！')
            break
    elif f==2:
        p1.beFight(p2._Person__fight)
        print('p1',end='')
        p1.printInfo()
        if p1._Person__blood<=0:
            print('p1没血了，凉了！')
            break
```

```python
class Person(object):
    def __init__(self,blood,gongji,fangyu):
        self.__blood=blood
        self.__gongji=gongji
        self.__fangyu=fangyu
    def beFighted(self,shanghai):
        self.__blood-=shanghai-self.__fangyu  #剩余的血量
    def getGongji(self):
        return self.__gongji
    def getBlood(self):
        return self.__blood
person1=Person(20,5,1)
person2=Person(30,4,2)
gongji_1=person1.getGongji()    #第一个人的攻击力
gongji_2=person2.getGongji()    #第二个人的攻击力

while True:
    num=input('攻击方：')
    if num=='1':
        person2.beFighted(gongji_1)
    elif num=='2':
        person1.beFighted(gongji_2)
    print('拳击手血剩余{},拳击手2血剩余{}'.format(person1.getBlood(),person2.getBlood()))
    if person1.getBlood()<=0:
        print('拳击手1挂了')
        break
    if person2.getBlood()<=0:
            print('拳击手2挂了')
            break
```

```python
#公司月末统计工资表，每个人的工资组成包括基本工资、奖金、扣除的五险一金=基本工资*10%，扣除的个人所得税
#五险一金除了个人扣除部分外，公司还需要为每个人交其基本工资*10%

#要求构造类保存每个人的工资信息，和公司需要支付的总钱数,五险一金数（公司交部分，不包括个人扣除部分），全体员工个人所得税总数
#能够打印出每个人的工资条，和公司支付的总钱数，五险一金数，全体员工个人所得税总数
#其中每个人的姓名，基本工资，奖金由文件读入，格式如下：
#张三，10000,1000
#李四，15000，1000
#王五，8000，500
class Wages(object):
    __all=0
    __allWu=0
    __allshui=0
    def __init__(self,name,jbgongzi,jiangjin,shui,a):
        self.__a=a
        self.__shui=shui
        self.__name=name
        self.__jbgongzi=jbgongzi
        self.__jiangjin=jiangjin
        self.__kcWu=self.__jbgongzi/10
        self.__kcGe=(self.__jbgongzi+self.__jiangjin-self.__kcWu-3500)*self.__shui-self.__a
        Wages.__all+=self.__jbgongzi*0.1+self.__jbgongzi+self.__jiangjin
        Wages.__allWu+=self.__kcWu
        Wages.__allshui+=self.__kcGe
    def printInfo(self):
        print('姓名：{},基本工资{}、奖金{}、扣除的五险一金{}，扣除的个人所得税{}'.format(self.__name,self.__jbgongzi,self.__jiangjin,self.__kcWu,self.__kcGe))
    @classmethod
    def printOutfo(cls):
        print('公司支付的总钱数{}，五险一金数{}，全体员工个人所得税总数{}'.format(Wages.__all,Wages.__allWu,Wages.__allshui))
f=open(r'C:\Users\Administrator\Desktop\Wages.txt','r')
b=f.readlines()
c1=b[0].split(',')
c2=b[1].split(',')
c3=b[2].split(',')
def If(c):
        if c<1500:
            shui,a=0.03,0
        elif c<4500:
            shui,a=0.1,105
        elif c<9000:
            shui,a=0.2,555
        elif c<35000:
            shui,a=0.25,1005
        elif c<55000:
            shui,a=0.3,2755
        elif c<80000:
            shui,a=0.35,5505
        else:
            shui,a=0.45,13505 
        return shui,a
shui1,a1=If(int(c1[1])*0.9-3500)
shui2,a2=If(int(c2[1])*0.9-3500)
shui3,a3=If(int(c3[1])*0.9-3500)
person1=Wages(c1[0],int(c1[1]),int(c1[2]),shui1,a1)
person2=Wages(c2[0],int(c2[1]),int(c2[2]),shui2,a2)
person3=Wages(c3[0],int(c3[1]),int(c3[2]),shui3,a3)
person1.printInfo()
person2.printInfo()
person3.printInfo()
Wages.printOutfo()
f.close()
        
        
```
