```python
import math
print(math.sin(0.5))       #使用math里的函数,math.不能省略


import random
x=random.random()             #获得0~1之间的随机数
y=x*100                      #获得1~100之间的随机数
print(y)
dir(random)
# help(random)
```

```python
from math import sin,cos    #通过from...import...方式引入某个函数
print(sin(0.5))             #再调用时，不用写math
print(cos(0.5))

from random import *       #引入random模块所有方法
```

```python
import math as t           #给math模块起了个别名，叫t
print(t.sin(0.5))

from math import sin as s   #引入math.sin方法，并给它起别名叫s，不建议使用
print(s(0.5))
```
