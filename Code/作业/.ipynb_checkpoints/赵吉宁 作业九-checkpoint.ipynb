```python
# 作业：
# 完成一个购物车的小系统
# 商城里有商品至少三种，通过列表保存商品的信息，包括编号（用数组下标即可，如0,1,2,3），商品名，价格
# 购物车功能包括增加商品，修改商品数量，删除商品，统计总价格四个函数
# 购物车里的有效商品用列表或字典保存，（参考购物车里用字典保存商品的编号和数量）
# 主程序中向用户展示所有商品，由用户输入商品编号和数量进行购买

goods=[{'编号:':0,'商品名:':'农夫山泉','价格:':2},{'编号:':1,'商品名:':'火腿肠','价格:':3},{'编号:':2,'商品名:':'棒棒糖','价格:':1}]
car={}
class Car(object):
    def addGoods(self,no,num):
        self.__no=no
        self.__num=num
        car[self.__no]=self.__num
    def changeGoodsNum(self,no,num):
        self.__no=no
        self.__num=num
        car[self.__no]=self.__num
    def delectGoods(self,no):
        self.__no=no
        del car[self.__no]
    def price(self):
        sum1=0
        for i,j in list(car.items()):
            sum1+=j*goods[i]['价格:']
        print('总价钱为：',sum1)   
print('商品如下：')
for i in goods:          
    print(i)        
people=Car()
print('操作1：购买商品\n操作2：更改商品数量\n操作3：删除商品\n操作0：结算商品总价钱')
m=int(input('请输入想要执行的操作：'))
n=1
while n==1:
    while m==1:
        a=int(input('请输入想要购买商品的编号：'))
        if a not in range(len(goods)):
            print('没有该商品！')
            print('当前购物车里的商品及数量为：',car)
            m=int(input('请输入想要执行的操作：'))
        else:
            b=int(input('请输入想要购买的数量：'))
            people.addGoods(a,b)
            c=input('还想继续买吗？T/F')
            if c=='F':
                print('当前购物车里的商品及数量为：',car)
                m=int(input('请输入想要执行的操作：'))
    while m==2:
        d=int(input('请输入想要更改的商品编号：'))
        if d not in list(car):
            print('购物车中没有该商品！')
            print('当前购物车里的商品及数量为：',car)
            m=int(input('请输入想要执行的操作：'))
        else:       
            e=int(input('请输入想要更改的商品数量：'))
            people.changeGoodsNum(d,e)
            f=input('还想继续更改吗？T/F')
            if f=='F':
                print('当前购物车里的商品及数量为：',car)
                m=int(input('请输入想要执行的操作：'))
    while m==3:
        g=int(input('请输入想要删除的商品编号：'))
        if g not in list(car):
            print('购物车中没有该商品！')
            print('当前购物车里的商品及数量为：',car)
            m=int(input('请输入想要执行的操作：'))
        else:
            people.delectGoods(g)
            h=input('还想继续删除吗？T/F')
            if h=='F':
                print('当前购物车里的商品及数量为：',car)
                m=int(input('请输入想要执行的操作：'))
    while m==0:
        people.price()
        n=0
        break 
```
