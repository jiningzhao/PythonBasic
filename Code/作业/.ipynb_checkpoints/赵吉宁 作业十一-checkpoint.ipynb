```python
# 1. 创建一个4*4数组，元素值从1~16，
#    1) 将该数组纵向分割成两个数组A、B，
#    2) 再组合生成一个新数组形式为[B A B]
import numpy as np
a=np.arange(1,17,1).reshape(4,4)
print(a)
(A,B)=np.vsplit(a,2)
print(A)
print(B)
C=np.hstack((B,A,B))
print(C)
```

```python
# 2. 创建一个3*4的数组，元素服从均匀分布，
#    1) 沿横轴计算数组的和、均值、标准差、方差、最小值、最大值、所有元素的和、所有元素的积。
#    2) 对数组按纵轴进行排序，打印结果
import random
a=np.random.uniform(1,10,size=12).reshape(3,4)
print('元素服从均匀分布的3*4的数组：\n',a)
sum1=a.sum(axis=1)
print('沿着横轴计算数组的和：\n',sum1)
mean1=a.mean(axis=1)
print('沿着横轴计算数组的均值：\n',mean1)
std1=a.std(axis=1)
print('沿着横轴计算数组的标准差：\n',std1)
var1=a.var(axis=1)
print('沿着横轴计算数组的方差：\n',var1)
min1=a.min(axis=1)
print('沿着横轴计算数组的最小值：\n',min1)
max1=a.max(axis=1)
print('沿着横轴计算数组的最大值：\n',max1)
cumsum1=a.cumsum()
print('计算数组的所有元素的和：\n',cumsum1[-1])
cumprod1=a.cumprod()
print('计算数组的所有元素的积：\n',cumprod1[-1])
a.sort(axis=0)
print('纵轴排序结果：\n',a)
```

```python
# 3. 创建两个2*3的矩阵A、B，元素为5~10之间的随机浮点数，
#    1) 计算两个矩阵相加、相减、相乘、对应元素相乘、相除的值
A=np.random.uniform(5,10,size=[6]).reshape(2,3)
B=np.random.uniform(5,10,size=[6]).reshape(2,3)
print('1)','-'*30)
print('A的值为：\n',A)
print('B的值为：\n',B)
print('两个矩阵相加的值为：\n',A+B)
print('两个矩阵相减的值为：\n',A-B)
print('两个矩阵相乘的值为：\n',A*B)
print('两个矩阵对应元素相乘的值为：\n',np.multiply(A,B))
print('两个矩阵相除的值的值为：\n',A/B)
#    2) 比较两个矩阵的大小，若A中所有元素都大于B，打印"A>B"，若A中任意元素大于B，打印"A_part > B"
print('2)','-'*30)
if (A>B).all():
    print('A>B')
elif (A>B).any():
    print('A_part>B')
#    3) 去掉两个矩阵中重复的元素   
print('3)','-'*30)
A_1=np.unique(A)
B_1=np.unique(B)
print(A_1,B_1)
#    4) 将矩阵A按行展开，将矩阵B按列展开
print('4)','-'*30)
A_2=A.flatten()
A_3=A.ravel()
B_2=B.flatten('F')
print(A_2)
print(A_3)
print(B_2)
```
